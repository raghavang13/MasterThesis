**In this notebook, I illustrate the variety of abundance profiles a user can set using class Abundances**

```python
import sys
sys.path.append('/home/raghav/sunbather/src/') #add your path to /sunbather/src/ here

import tools
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Different profile shapes**

```python
solar_ab = tools.Abundances()
```

```python
ellipse_ab = tools.Abundances()
ellipse_ab.set_fractionationprofile(elements='C', proftype='ellipse', frac_intervals=[1.0,2.0], vals=0.1)
```

```python
plaw_ab = tools.Abundances()
plaw_ab.set_fractionationprofile(elements='C', proftype='powerlaw', frac_intervals=[1.0,2.0], vals=0.1)
```

```python
strline_ab = tools.Abundances()
strline_ab.set_fractionationprofile(elements='C', proftype='strline', frac_intervals=[1.0,2.0], vals=0.1)
```

```python
parabola_ab = tools.Abundances()
parabola_ab.set_fractionationprofile(elements='C', proftype='parabola', frac_intervals=[1.0,2.0], vals=0.1, parabola_vertices=[1.3])
```

```python
expdecay_ab = tools.Abundances()
expdecay_ab.set_fractionationprofile(elements='C', proftype='expdecay', frac_intervals=[1.0,2.0], vals=0.1)
```

```python
fig, ax = plt.subplots(1)
ax.plot(solar_ab.abundance_profiles.index.values, solar_ab.abundance_profiles['C'], color='xkcd:deep sky blue', label='Solar, no fractionation', linewidth=2)
ax.plot(plaw_ab.abundance_profiles.index.values, plaw_ab.abundance_profiles['C'], color='xkcd:aqua green', label='Power law', linewidth=2)
ax.plot(ellipse_ab.abundance_profiles.index.values, ellipse_ab.abundance_profiles['C'], color='xkcd:tangerine', label='Ellipse', linewidth=2)
ax.plot(expdecay_ab.abundance_profiles.index.values, expdecay_ab.abundance_profiles['C'], color='xkcd:light magenta', label='Exponential decay', linewidth=2)
ax.plot(strline_ab.abundance_profiles.index.values, strline_ab.abundance_profiles['C'], color='xkcd:bordeaux', label='Straight line', linewidth=2)
ax.plot(parabola_ab.abundance_profiles.index.values, parabola_ab.abundance_profiles['C'], color='xkcd:grey blue', label='Parabola', linewidth=2)
tools.set_alt_ax(ax, altmax=4)
ax.set_ylabel('Mixing Ratio', size=14)
ax.set_xlabel('Radius [R$_P$]', size=14)
plt.yticks(size=13)
plt.xticks(size=13)
ax.legend(prop={'size':12})
#plt.savefig('Methods_ProfShapes.png', dpi=400, bbox_inches='tight')
plt.show()
```

**Different mixing ratios and bulk compositions**

```python
solar_ab = tools.Abundances()
z10_ab = tools.Abundances()
z10_ab.set_metallicity(2.)
solar_frac1_ab = tools.Abundances()
solar_frac1_ab.set_fractionationprofile(elements='C', proftype='powerlaw', frac_intervals=[1.0,2.0], vals=0.1, fraction=True)
solar_frac2_ab = tools.Abundances()
solar_frac2_ab.set_fractionationprofile(elements='C', proftype='powerlaw', frac_intervals=[1.0,2.0], vals=0.2, fraction=True)
z10_frac1_ab = tools.Abundances()
z10_frac1_ab.set_metallicity(2.)
z10_frac1_ab.set_fractionationprofile(elements='C', proftype='powerlaw', frac_intervals=[1.0,2.0], vals=0.2,fraction=True)
```

```python
fig, ax = plt.subplots(1)
ax.plot(solar_ab.abundance_profiles.index.values, solar_ab.abundance_profiles['C'], color='xkcd:deep sky blue', label='Solar, no fractionation', linewidth=2)
ax.plot(solar_frac1_ab.abundance_profiles.index.values, solar_frac1_ab.abundance_profiles['C'], color='xkcd:tangerine', label='Solar, final value 10%', linewidth=2)
ax.plot(solar_frac2_ab.abundance_profiles.index.values, solar_frac2_ab.abundance_profiles['C'], color='xkcd:light magenta', label='Solar, final values 20%', linewidth=2)
ax.plot(z10_ab.abundance_profiles.index.values, z10_ab.abundance_profiles['C'], color='xkcd:aqua green', label='Metallicity 2, no fractionation', linewidth=2)
ax.plot(z10_frac1_ab.abundance_profiles.index.values, z10_frac1_ab.abundance_profiles['C'], color='xkcd:bordeaux', label='Metallicity 2, final value 20%', linewidth=2)
ax.set_ylabel('Mixing Ratio', size=14)
ax.set_xlabel('Radius [R$_P$]', size=14)
plt.yticks(size=13)
plt.xticks(size=13)
ax.legend(prop={'size':11}, loc=(0.4, 0.6))
tools.set_alt_ax(ax, altmax=4)
#plt.savefig('Methods_MixRats.png', dpi=400, bbox_inches='tight')
plt.show()
```

**Different fractionation intervals**

```python
solar_ab = tools.Abundances()
fracint1_ab = tools.Abundances()
fracint1_ab.set_fractionationprofile(elements='C', proftype='powerlaw', frac_intervals= [1.0, 2.0], vals=0.1,fraction=True)
fracint2_ab = tools.Abundances()
fracint2_ab.set_fractionationprofile(elements='C', proftype='powerlaw', frac_intervals= [1.2, 2.5], vals=0.1,fraction=True)
fracint3_ab = tools.Abundances()
fracint3_ab.set_fractionationprofile(elements='C', proftype='powerlaw', frac_intervals= [1.5, 4.0], vals=0.1,fraction=True)
```

```python
fig, ax = plt.subplots(1)
ax.plot(solar_ab.abundance_profiles.index.values, solar_ab.abundance_profiles['C'], color='xkcd:deep sky blue', label='Solar, no fractionation', linewidth=2)
ax.plot(fracint1_ab.abundance_profiles.index.values, fracint1_ab.abundance_profiles['C'], color='xkcd:tangerine', label='Fractionation interval 1-2R$_P$', linewidth=2)
ax.plot(fracint2_ab.abundance_profiles.index.values, fracint2_ab.abundance_profiles['C'], color='xkcd:light magenta', label='Fractionation interval 1.2-2.5R$_P$', linewidth=2)
ax.plot(fracint3_ab.abundance_profiles.index.values, fracint3_ab.abundance_profiles['C'], color='xkcd:aqua green', label='Fractionation interval 1.5-4R$_P$', linewidth=2)
ax.set_ylabel('Mixing Ratio', size=14)
ax.set_xlabel('Radius [R$_P$]', size=14)
plt.yticks(size=13)
plt.xticks(size=13)
ax.legend(prop={'size':11}, loc=(0.4, 0.6))
tools.set_alt_ax(ax, altmax=5)
#plt.savefig('Methods_FracInt.png', dpi=400, bbox_inches='tight')
plt.show()
```
